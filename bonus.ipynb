```python
#from sqlalchemy import create_engine
import psycopg2 as pg
import pandas.io.sql as psql
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
import numpy as np
connection = pg.connect("host=localhost dbname=Homework user=test password=test")
salaries = psql.read_sql('SELECT * FROM salaries', connection)
#Average Salary is 52970.732451403885
mean_salary = np.mean(salaries.salary)
print(mean_salary)
plt.xlabel('Salary ($)')
plt.ylabel('Counts')
plt.hist(salaries.salary)

# Demonstrate calculating the variance and standard deviation using the different modules
var_numpy = np.var(salaries.salary,ddof = 0)
print(f"The population variance using the NumPy module is {var_numpy}")

sd_numpy = np.std(salaries.salary,ddof = 0)
print(f"The population standard deviation using the NumPy module is {sd_numpy}")

print(f"Roughly 68% of the data is between {round(mean_salary-sd_numpy,3)} and {round(mean_salary+sd_numpy,3)}")
print(f"Roughly 95% of the data is between {round(mean_salary-2*sd_numpy,3)} and {round(mean_salary+2*sd_numpy,3)}")
print(f"Roughly 99.7% of the data is between {round(mean_salary-3*sd_numpy,3)} and {round(mean_salary+3*sd_numpy,3)}")
```

```python
#bar chart
title_salary = psql.read_sql('select t.title,round(avg(s.salary),2) average_salary from employees e, titles t, salaries s where e.emp_title_id = t.title_id and e.emp_no = s.emp_no group by t.title ', connection)
x_axis = np.arange(len(title_salary))
plt.bar(title_salary.title, title_salary.average_salary, color='r', alpha=0.5, align="center") 
plt.title("Titles vs Salaries")
plt.xticks(rotation='vertical')
plt.xlabel("Title")
plt.ylabel("Salary ($)")
```
